```python
table1 = pd.pivot_table(alive_25_students_df, values='studentID', index=['Rural/Urban Status'],
                       columns=None, aggfunc="count")
table1
```

```python
table2 = pd.pivot_table(merged_df, values='ano', index=['Rural/Urban Status'],
                       columns=None, aggfunc="count")
table2
```

```python
merged_df2 = pd.merge(table1,  
                     table2,  
                     on ='Rural/Urban Status',  
                     how ='left').reset_index()
merged_df2
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Define bar width
bar_width = 0.35

# Set positions of bars on x-axis
r1 = range(len(merged_df2['Rural/Urban Status']))
r2 = [x + bar_width for x in r1]

# Create bars
bars1 = plt.bar(r1, merged_df2['studentID'], color='blue', width=bar_width, edgecolor='grey', label='Students')
bars2 = plt.bar(r2, merged_df2['ano'], color='red', width=bar_width, edgecolor='grey', label='Accidents')

# Add labels and title
plt.xlabel('County Type', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(merged_df2['Rural/Urban Status']))], merged_df2['Rural/Urban Status'])
plt.ylabel('Collisions')
plt.title('Number of Students and Accidents per County Type')

# Add value labels on top of the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # va: vertical alignment

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # va: vertical alignment

# Add legend
plt.legend()

# Show plot
plt.show()
```
